```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv('../../doc1/Data/raw_analyst_ratings/raw_analyst_ratings.csv')
```

```python

publisher_counts = df['publisher'].value_counts()
plt.figure(figsize=(12, 6))
publisher_counts.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Publishers by Article Count')
plt.xlabel('Publisher')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python

df['domain'] = df['publisher'].str.split('@').str[-1]

domain_counts = df['domain'].value_counts()

plt.figure(figsize=(12, 6))
domain_counts.head(10).plot(kind='bar', color='lightcoral')
plt.title('Top 10 Domains by Article Count')
plt.xlabel('Domain')
plt.ylabel('Number of Articles')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
